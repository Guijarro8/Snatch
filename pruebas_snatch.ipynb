```python
def get_daily_EOD(company):
    params = {"symbols": company, "limit": "1000"}
    r = requests.get(MARKETSTACK_URL_EOD, params=params)
    data = r.json()
    df = pd.json_normalize(data["data"])

    return df
```

```python
def get_tickers(stock_exchange):
    params = {"limit": "1000", "exchange": stock_exchange}
    r = requests.get(MARKETSTACK_URL_TICKERS, params=params)
    data = r.json()
    print(r.status_code)
    df = pd.json_normalize(data["data"])

    return df
```

```python
def get_exchange():
    params = {"limit": "1000"}
    r = requests.get(MARKETSTACK_URL_EXCHANGES, params=params)
    data = r.json()
    print(r.status_code)
    df = pd.json_normalize(data["data"])

    return df
```
